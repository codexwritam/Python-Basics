1. What is Tableau? Explain its importance in Business Intelligence and how it helps in data-driven decision-making.
- Tableau is a powerful data visualization and Business Intelligence (BI) tool that helps users analyze data and create interactive visual reports, charts, dashboards, and stories without deep programming knowledge.

Importance of Tableau in Business Intelligence:

- Converts raw data into meaningful insights

- Supports real-time data analysis

- Helps businesses track sales, profit, trends, and performance

- Easy drag-and-drop interface for faster analysis

Role in Data-Driven Decision-Making:

- Identifies patterns, trends, and outliers

- Enables comparison across regions, products, and time

- Helps management take accurate, evidence-based decisions

- Improves forecasting and strategic planning

2. Explain the role of the following Tableau components:
a) Data Pane b) Worksheet c) Dashboard d) Story
- a) Data Pane

The Data Pane is located on the left side of Tableau.
It displays:

- Dimensions

- Measures

- Calculated fields

- Parameters

It is used to drag fields into Rows, Columns, Filters, and Marks.

- b) Worksheet

A Worksheet is where individual visualizations (charts, graphs, maps) are created.
Each worksheet represents one visualization.

- c) Dashboard

A Dashboard combines multiple worksheets into a single view.
It helps users analyze data from different perspectives at once and supports interactivity using filters and actions.

- d) Story

A Story is a sequence of worksheets or dashboards used to explain data insights step-by-step like a presentation.

3. What is the difference between Dimensions and Measures in Tableau? Provide examples of each.
- | Dimension               | Measure               |
| ----------------------- | --------------------- |
| Qualitative data        | Quantitative data     |
| Used for categorization | Used for calculations |
| Blue color              | Green color           |
| Discrete by default     | Continuous by default |

Examples:

Dimensions: Country, Product, Category, Segment, Date

Measures: Sales, Profit, Units Sold, Quantity

4. Define and explain the purpose of Filters, Parameters, and Sets in
Tableau.
- Filters

Filters are used to limit data shown in a visualization.
Example: Filter sales data for the year 2014.

- Parameters

Parameters allow users to manually control values and interact with visualizations dynamically.
Example: Switch between Units Sold and Profit.

- Sets

Sets are custom groups of data based on conditions.
Example: Top 10 products by sales.
