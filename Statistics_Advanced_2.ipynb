1. What is hypothesis testing?
- A procedure to use sample data to decide between two competing claims about a population. We compute a test statistic, compare it to a reference distribution, and decide whether the sample provides enough evidence to reject the default claim.

2. Null vs. Alternative hypothesis

- Null hypothesis (H₀): the default/no-effect claim (e.g., μ = 50).

- Alternative hypothesis (H₁/Ha): what we want to show (e.g., μ ≠ 50, μ > 50, or μ < 50).
We either reject H₀ in favor of Ha, or fail to reject H₀ (we never “accept” H₀ with certainty).

3. Significance level (α)

- α is the allowed probability of wrongly rejecting a true H₀ (Type I error). Common choices are 0.05 or 0.01. If the p-value ≤ α, we reject H₀; otherwise we fail to reject.

4. Type I and Type II errors

- Type I (false positive): Reject H₀ when H₀ is true.
 - Example: Concluding a new drug works when it actually doesn’t.

- Type II (false negative): Fail to reject H₀ when Ha is true.
 - Example: Concluding a new drug doesn’t work when it actually does.

5.  Z-test vs T-test — when to use?

- Z-test: Population standard deviation (σ) is known or sample size is large and we’re comfortable using the normal approximation.
- T-test: σ is unknown and sample size is small; uses the t-distribution with

```python
 #6 rite a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
 import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)
samples = np.random.binomial(n=10, p=0.5, size=5000)

plt.figure()
plt.hist(samples, bins=range(0, 12), align='left', rwidth=0.9)
plt.title("Binomial(n=10, p=0.5) - 5,000 simulations")
plt.xlabel("Successes in 10 trials")
plt.ylabel("Frequency")
plt.show()
```

```python
#7 Implement hypothesis testing using Z-statistics for a sample dataset in
# Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
# 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
# 50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
import numpy as np
from statistics import mean, stdev
from math import sqrt
from math import erf, sqrt as msqrt

def norm_cdf(z):
    return 0.5 * (1 + erf(z / msqrt(2)))

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

xbar = mean(sample_data); s = stdev(sample_data); n = len(sample_data); mu0 = 50
z = (xbar - mu0) / (s / sqrt(n))
pval = 2 * (1 - norm_cdf(abs(z)))
print(z, pval)

```

```python
#8 Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(7)
data = np.random.normal(100, 15, 200)

xbar = np.mean(data)
s = np.std(data, ddof=1)
se = s / np.sqrt(len(data))
z = 1.96
ci = (xbar - z*se, xbar + z*se)
print(ci)

plt.figure()
plt.hist(data, bins=30)
plt.title("Simulated Normal Data (n=200)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

```

```python
#9 Write a Python function to calculate the Z-scores from a dataset and
# visualize the standardized data using a histogram. Explain what the Z-scores represent
# in terms of standard deviations from the mean.
import numpy as np
import matplotlib.pyplot as plt

def z_scores(arr):
    m = np.mean(arr); sd = np.std(arr, ddof=1)
    return (arr - m) / sd

zs = z_scores(np.array(sample_data))
print(np.mean(zs), np.std(zs, ddof=1))  # ≈ 0 and 1

plt.figure()
plt.hist(zs, bins=15)
plt.title("Histogram of Z-scores (standardized)")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()
```
