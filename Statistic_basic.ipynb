1. What is the difference between descriptive statistics and inferential
statistics? Explain with examples.

- The difference between descriptive statistics and inferential statistics:

Descriptive statistics and inferential statistics are two main branches of statistics that help us understand data.

- Descriptive Statistics:
Descriptive statistics is all about summarizing and describing the main features of a dataset. It helps us understand what the data looks like without making any conclusions beyond the data we have.

Examples:

If we collect the marks of 30 students in a class and calculate the average (mean), median, mode, or standard deviation, we are using descriptive statistics.

Making a bar graph or pie chart of students’ favorite subjects is also descriptive.

In simple words: It tells us “what” the data shows.

- Inferential Statistics:
Inferential statistics is about making predictions or generalizations about a larger group based on a smaller sample of data. It uses probability to make conclusions.

Examples:

If we survey 100 people from a city to guess the favorite food of the whole city, and we say, “Pizza is the most liked food in the city,” we are using inferential statistics.

Doing a hypothesis test or finding a confidence interval are part of inferential statistics.

2. What is sampling in statistics? Explain the differences between random
and stratified sampling?

- What is Sampling in Statistics?
Sampling is the process of selecting a small group (sample) from a larger group (population) to collect data and draw conclusions.
Since it is difficult or costly to study the entire population, we take a sample that represents the population.

Example:
If a school has 1000 students and we survey only 100 students about their favorite subject, that 100 is a sample.

Types of Sampling – Random vs Stratified
- Random Sampling:
In random sampling, every member of the population has an equal chance of being selected.

Example:
If we put the names of all 1000 students in a box and pick 100 names randomly, that’s random sampling.

- Advantages:

Simple and easy to understand

No bias

- Disadvantage:

Might not represent different groups in the population properly

- Stratified Sampling:
In stratified sampling, we divide the population into groups (called strata) based on a specific characteristic (like gender, age, class, etc.), and then take a random sample from each group.

Example:
If the school has 600 boys and 400 girls, we divide them into two groups and then randomly select 60 boys and 40 girls to maintain the same ratio.

- Advantages:

More accurate representation of the population.

Ensures all groups are included.

- Disadvantage:

More complex than simple random sampling.

3. Define mean, median, and mode. Explain why these measures of central
tendency are important.

- Definition of Mean, Median, and Mode:
These three are called measures of central tendency because they help us find the center or average value of a dataset.

- Mean (Average):
The mean is found by adding all the numbers in a dataset and then dividing by the total number of values.

Formula:


Mean=
Total number of values/
Sum of all values
​

Example:
If marks are: 70, 80, 90
Mean = (70 + 80 + 90) / 3 = 240 / 3 = 80

- Median:
The median is the middle value when the numbers are arranged in order.

If the number of values is odd, the median is the middle number.

If the number of values is even, the median is the average of the two middle numbers.

Example:
For values: 10, 20, 30 → Median = 20
For values: 10, 20, 30, 40 → Median = (20 + 30)/2 = 25

- Mode:
The mode is the number that appears most often in a dataset.

Example:
For values: 4, 5, 6, 4, 7 → Mode = 4 (because 4 appears twice)

A dataset can have no mode, one mode, or more than one mode.

Why These Measures Are Important:
- Summarize Data Easily:
They help us understand a large dataset by giving a single representative value.

- Compare Different Groups:
We can compare two sets of data using mean or median (e.g., average marks of two classes).

-  Used in Real Life:
Used in many areas like:

Business: average sales

Education: average test scores

Weather: average temperature

- Handle Different Types of Data:

Mean is useful when there are no extreme values.

Median is best when the data has outliers (extreme values).

Mode is helpful for categorical data (like most common shoe size, favorite color, etc.).

4.Explain skewness and kurtosis. What does a positive skew imply about the
data?

- Skewness and Kurtosis in Statistics:
Both skewness and kurtosis are used to describe the shape of a data distribution.

- Skewness:
Skewness tells us how symmetrical the data is.

If data is perfectly symmetrical, skewness is 0 (like a bell-shaped curve).

If the data is not symmetrical, it is skewed (either to the left or right).

- Types of Skewness:
Positive Skew (Right Skew):

The right tail is longer.

Most values are on the left, and few large values are on the right.

Mean > Median > Mode

Example: Income data — most people earn less, a few earn a lot.

Negative Skew (Left Skew):

The left tail is longer.

Most values are on the right, and few small values pull the tail to the left.

Mean < Median < Mode

-Kurtosis:
Kurtosis tells us about the "tailedness" or peakedness of the distribution — in simple words, how sharp or flat the peak is compared to a normal bell curve.

- Types of Kurtosis:
Mesokurtic: Normal distribution (kurtosis = 0)

Leptokurtic: Sharp peak, heavy tails (more outliers)

Platykurtic: Flat peak, light tails (fewer outliers)

What Does Positive Skew Mean?
A positive skew means:

The tail on the right side is longer.

Most data values are smaller, but there are a few large values that stretch the tail.

Mean is greater than median.

Example: Marks of a difficult exam where most students scored low, but a few scored very high.

















```python
#5. Implement a Python program to compute the mean, median, and mode ofa given list of numbers.
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

import statistics

# Given list of numbers
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate Mean
mean = statistics.mean(numbers)

# Calculate Median
median = statistics.median(numbers)

# Calculate Mode
mode = statistics.mode(numbers)

# Print Results
print("Given Numbers:", numbers)
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
```

```python
#6.Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

import numpy as np

# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to numpy arrays
x = np.array(list_x)
y = np.array(list_y)

# Calculate Covariance
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0][1]

# Calculate Correlation Coefficient
correlation_matrix = np.corrcoef(x, y)
correlation = correlation_matrix[0][1]

# Print results
print("List X:", list_x)
print("List Y:", list_y)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)

```

```python
#7.  Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

import matplotlib.pyplot as plt
import numpy as np

# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Create the boxplot
plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Boxplot of Given Data")
plt.xlabel("Value")
plt.grid(True)
plt.show()

# Identify outliers manually using IQR method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Display statistics
print("Q1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR:", IQR)
print("Lower Bound:", lower_bound)
print("Upper Bound:", upper_bound)
print("Outliers:", outliers)
```

```python
#8. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.i.Explain how you would use covariance and correlation to explore this relationship.ii.Write Python code to compute the correlation between the two lists:
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

import numpy as np

# Data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert to NumPy arrays
x = np.array(advertising_spend)
y = np.array(daily_sales)

# Covariance
cov_matrix = np.cov(x, y, bias=False)
covariance = cov_matrix[0][1]

# Correlation Coefficient
correlation_matrix = np.corrcoef(x, y)
correlation = correlation_matrix[0][1]

# Output results
print("Advertising Spend:", x)
print("Daily Sales:", y)
print("Covariance:", covariance)
print("Correlation Coefficient:", correlation)
```

```python
#9. : Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.i.Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.ii. Write Python code to create a histogram using Matplotlib for the survey data:
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

import matplotlib.pyplot as plt
import numpy as np
import statistics

# Customer satisfaction survey scores
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Summary Statistics
mean = statistics.mean(survey_scores)
median = statistics.median(survey_scores)
mode = statistics.mode(survey_scores)
std_dev = statistics.stdev(survey_scores)

# Print summary statistics
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Standard Deviation:", round(std_dev, 2))
print("Min:", min(survey_scores))
print("Max:", max(survey_scores))

# Create Histogram
plt.hist(survey_scores, bins=range(1, 12), edgecolor='black', color='skyblue', align='left')
plt.title('Customer Satisfaction Score Distribution')
plt.xlabel('Satisfaction Score (1–10)')
plt.ylabel('Number of Customers')
plt.xticks(range(1, 11))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
