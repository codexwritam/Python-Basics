1. What is the difference between 'Paste' and 'Paste Special' in Excel? Briefly
explain with examples.
- In Excel, Paste and Paste Special are used to insert copied data, but they differ in flexibility and control.

Paste: This option pastes everything from the copied cell, including:

- Values

- Formulas

- Formatting

- Comments

Example:
If a cell contains a formula =A1+B1 with formatting, using Paste will paste the formula along with formatting.

Paste Special:
This option allows users to paste specific elements only, such as:

- Values

- Formulas

- Formats

- Transpose

- Operations (Add, Subtract, etc.)

Example:
If a cell has a formula and you use Paste Special → Values, only the calculated result will be pasted, not the formula.

Conclusion:
Paste Special provides more control and is useful when working with cleaned or transformed data.

4. What are logical functions in Excel? Provide examples of at least two
logical functions and their applications.
- Logical functions are used to test conditions and return results based on TRUE or FALSE values.

1. IF Function

Syntax:
=IF(condition, value_if_true, value_if_false)

Example:
=IF(A1>=40,"Pass","Fail")
Used to determine exam results.

2. AND Function

Syntax:
=AND(condition1, condition2)

Example:
=AND(A1>50, B1>50)
Returns TRUE only if both conditions are met.

Applications:
Used in grading systems, eligibility checks, and decision-making models.

5. Discuss the purpose of 'XLOOKUP' and how it differs from the traditional
'VLOOKUP' function.
- XLOOKUP is a modern lookup function that replaces VLOOKUP and HLOOKUP.

    Purpose of XLOOKUP

  - Finds values in a range and returns matching results.

  - Works vertically and horizontally.

Differences Between XLOOKUP and VLOOKUP:
| Feature                | XLOOKUP | VLOOKUP            |
| ---------------------- | ------- | ------------------ |
| Search direction       | Any     | Left to right only |
| Column index required  | No      | Yes                |
| Handles missing values | Yes     | No                 |
| Default match          | Exact   | Approximate        |
