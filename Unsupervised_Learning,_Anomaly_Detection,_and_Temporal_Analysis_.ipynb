1. What is Dimensionality Reduction? Why is it important in machine
learning?
- Dimensionality Reduction is the process of reducing the number of input features or variables in a dataset while retaining as much relevant information as possible. It transforms high-dimensional data into a lower-dimensional form.

- Importance:

  - Improves Model Performance: Reduces computational cost and overfitting.

  - Simplifies Visualization: Allows visualization of complex data in 2D or 3D space.

  - Removes Redundancy: Eliminates correlated or irrelevant features.

  - Enhances Learning Speed: Models train faster with fewer dimensions.

2. Name and briefly describe three common dimensionality reduction
techniques.
- Principal Component Analysis (PCA):
Converts correlated features into a smaller set of uncorrelated variables called principal components based on variance.

- Linear Discriminant Analysis (LDA):
Finds the feature space that maximizes class separability. Commonly used for classification tasks.

- t-Distributed Stochastic Neighbor Embedding (t-SNE):
A nonlinear technique used for visualization by preserving local similarities between data points in lower dimensions.

3. What is clustering in unsupervised learning? Mention three popular
clustering algorithms.
- Clustering is an unsupervised learning technique that groups similar data points into clusters based on their characteristics, without predefined labels.

- Popular Algorithms:

  - K-Means Clustering: Divides data into K clusters by minimizing the within-cluster variance.

  - Hierarchical Clustering: Builds a hierarchy of clusters through agglomerative or divisive methods.

  - DBSCAN (Density-Based Spatial Clustering of Applications with Noise): Groups data points based on density and identifies outliers as noise.

4. Explain the concept of anomaly detection and its significance.
- Anomaly Detection identifies data points that deviate significantly from the expected pattern in a dataset.

- Significance:

- Fraud Detection: Detects unusual transactions or behaviors.

- System Monitoring: Identifies irregularities in network or system performance.

- Safety and Security: Detects equipment malfunctions or cyber-attacks.

- Quality Control: Spots defective items in manufacturing.

- Distance/Clustering-Based Methods:

6.  What is time series analysis? Mention two key components of time series
data.
- Time Series Analysis involves studying data points collected or recorded at regular time intervals to identify patterns, trends, and dependencies over time.

- Key Components:

  - Trend: The long-term upward or downward movement in data.

  - Seasonality: Repeating short-term patterns occurring at fixed intervals (e.g., monthly or yearly).

7. Describe the difference between seasonality and cyclic behavior in time
series.
- Seasonality:
Refers to periodic fluctuations that occur at fixed intervals (e.g., increased sales during holidays).
Duration and frequency are predictable.

- Cyclic Behavior:
Refers to fluctuations that occur over irregular time intervals due to economic or natural factors.
Duration is not fixed and depends on underlying conditions.

```python
# Write Python code to perform K-means clustering on a sample dataset.
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=200, centers=3, cluster_std=1.0, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

centers = kmeans.cluster_centers_
labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50)
plt.scatter(centers[:, 0], centers[:, 1], c='red', marker='x', s=200)
plt.title("K-Means Clustering Example")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

10. How can time series analysis be used for anomaly detection?
- Time series analysis can be used to detect anomalies by identifying points that deviate from expected temporal patterns.

Methods:

- Forecasting Models (e.g., ARIMA, LSTM):
Predict future values and flag large deviations as anomalies.

- Statistical Thresholding:
Uses moving averages or confidence intervals to detect unusual spikes or drops.

- Seasonal Decomposition:
Separates trend, seasonality, and residuals; anomalies appear as irregular residuals.

Applications:
Used in network monitoring, stock market analysis, and detecting equipment failures over time.
